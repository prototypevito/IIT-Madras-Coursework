### Calculating Pearson factor, mean range and variance range

```python
def get_ans(data,sample_data):
    import numpy as np 
    from scipy.stats import chi2,t,moment
    import seaborn as sns
    import matplotlib.pyplot as plt
    sns.set_style('darkgrid')
    
    data = np.array(data)
    sample_data = np.array(sample_data)
    
    
    # Calculating Pearson coeff using moments.
    # NOTE: Using the skew and kurtosis functions from scipy.stats gives a very different answer.
    moments = [moment(data, i) for i in range(1,5)]
    beta1 = moments[2]**2/moments[1]**3
    beta2 = moments[3]/moments[1]**2
    ans1 = 6+3*beta1-2*beta2
    
    # Mean range using student t distribution
    std = sample_data.std()
    mean = sample_data.mean()
    p = t.ppf(0.995,99)
    mean_range = -p*std/10**0.5 + mean, p*std/10**0.5 + mean
    
    # Variance range using chi sqaured distribution. Ref: http://www.milefoot.com/math/stat/ci-variances.htm
    var_range = (99*sample_data.var()/chi2.ppf(0.995,99)*10,99*sample_data.var()/chi2.ppf(0.005,99)*10)
    
    print(f'Pearson Coeff = {ans1}\n\nMean Range = {mean_range}\n\nVariance Range = {var_range}\n\n')
```

```python
get_ans(data,sample_data)
```
